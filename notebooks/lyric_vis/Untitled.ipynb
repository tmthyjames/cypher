```python
tfidf_bank = {}
for row in tfidf[tfidf.word.astype(str).str.isalnum()].itertuples():
    if row.ranker_genre not in tfidf_bank:
        tfidf_bank[row.ranker_genre] = {}
    tfidf_bank[row.ranker_genre][row.word] = {
        'tfidf': row.tf_idf,
        'tf': row.tf,
        'idf': row.idf,
        'n': str(row.n)
    }
```
